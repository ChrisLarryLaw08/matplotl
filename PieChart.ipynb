```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
ds = pd.read_csv('albprecip.csv')
```

```python
month = ds.Month
precip = ds.Precip
explode = np.zeros(12)
explode[0] += 0.1
explode[6] += 0.1
```

```python
fig = plt.figure(figsize=(11,8.5))
ax = fig.add_subplot(1,1,1)
ax.set_title("Monthly Precipitation at Albany NY")
ax.set_facecolor('yellow')
#ax.pie(precip,labels=month);
wedges, labels, autopct = ax.pie(precip,labels=month,explode=explode,startangle=90,counterclock=False,shadow=True,autopct='%1.1f%%',textprops = {'fontsize':8},wedgeprops={'edgecolor':'black'});
plt.setp(labels,fontsize=14)
```

```python
ax.tight_layout()
```
